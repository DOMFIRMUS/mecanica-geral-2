```

import time
import math as m
from prettytable import PrettyTable

massa_a = 0.35
massa_b = 0.7
Va = 6
Vb = 0
angulo = 20
e = 0
##### MODIFICA O e ######## A)e=1 B)e=0.6 C)e=0.75 D)e=0  ################################

tabela = PrettyTable()
tabela.title = "" 
tabela.field_names = ["Theta","e","Va","Vb","V'a","V'b","Diferença de V","%Energia"]

Vrelativa = []
theta = []
EnergiaPerda = []
VaLinha = []
VbLinha = []


def va_l(angulo, e):
  return (6*(m.sin(m.radians(angulo)))**2 - 12*e) / (2 + (m.sin(m.radians(angulo)))**2)

def vb_l(angulo, e):
  return (6*(m.sin(m.radians(angulo)))*(1+e)) / (2 + (m.sin(m.radians(angulo)))**2)

def ener(massa_a,massa_b,va,valinha,vblinha):
  return 100*abs(((massa_a*((valinha)**2-(va)**2)) + (massa_b*(vblinha)**2))/(massa_a*(va)**2))

while angulo<=150:
  
  vla = va_l(angulo, e)
  vlb = vb_l(angulo, e)
  energia = round(ener(massa_a,massa_b,Va,vla,vlb),3)
  
  Vr = vlb - vla

  tabela.add_row([round(angulo,3),round(e,3),round(Va,3),round(Vb,3),round(vla,3),round(vlb,3),round(Vr,3),energia])

  theta.append(angulo)
  VaLinha.append(vla)
  VbLinha.append(vlb)
  Vrelativa.append(Vr)
  EnergiaPerda.append(energia)

  angulo += 5
 

print(tabela)
```

```
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12, 4))
plt.plot( theta, Vrelativa,'o-', color='red', label = 'Vr')
plt.xlabel('Ângulos (graus)')
plt.title('Resultado calculado para os ângulos')
plt.ylabel('Velocidade relativa (m/s)')
plt.legend(loc='lower left')
plt.grid('True')
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(8, 4))
plt.plot( theta, VaLinha,'o-', color='purple', label = 'Va_linha')
plt.plot( theta, VbLinha,'o-', color='orange', label = 'Vb_linha')
plt.xlabel('Ângulos (graus)')
plt.ylabel('Velocidade (m/s)')
plt.title('Resultado calculado para os ângulos')
plt.legend(loc='lower center')
plt.grid('True')
plt.show()
```
