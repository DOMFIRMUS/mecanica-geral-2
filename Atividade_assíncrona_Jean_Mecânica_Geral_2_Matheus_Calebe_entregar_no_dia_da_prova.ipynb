```
# Método para encontrar as velocidades V'a e V'b e a percentagem da energia perdida
import time
import math as m
from prettytable import PrettyTable
massaa = 0.350 #Kg
massab = 0.700 #Kg
Va = 6.000 #m/s Vo=Va
Vb = 0.000 #m/s Vb=0 Pois está no repouso
angulo = 20 # Ângulo inicial vai de 20º até 150º, iterando 5º.
e = 0.000 # a) e=1.000 , b) e=0.600 c) e=0.750 d)e=0.000   
# Para importar as bibliotecas necessárias.
tabela = PrettyTable()
# Criando uma tabela.
tabela.title = f'Método para theta = {e}'
tabela.field_names = ["Ângulo","e","Va","Vb","V'a","V'b","Vrelativa","Energia Perdida","T(s)"]
# Montando o cabeçalho da tabela.
Angulo = []
va_Linha = []
vb_Linha = []
v_relativa = []
perda_de_energia = []
# Serão usados para o armazenamento e plotagem dos dados.
def va_linha(angulo, e):
  return (6*(m.sin(m.radians(angulo)))**2 - 12*e) / (2 + (m.sin(m.radians(angulo)))**2)
def vb_linha(angulo, e):
  return (6*(m.sin(m.radians(angulo)))*(1+e)) / (2 + (m.sin(m.radians(angulo)))**2)
def energia(massa_a,massa_b,va,valinha,vblinha):
  return 100*abs(((massa_a*((valinha)**2-(va)**2)) + (massa_b*(vblinha)**2))/(massa_a*(va)**2))
# Declarando as equações encontradas.
def velocidades(massaa,massab,Va,Vb,angulo,e):
  while angulo<=150:
    inicio = time.time()
    # Iniciando o contador temporal.
    VLA = va_linha(angulo, e)
    VLB = vb_linha(angulo, e)
    Energia = round(energia(massaa,massab,Va,VLA,VLB),6)
    # Aplicando os métodos.
    fim = time.time()
    # Finalizando o contador temporal.
    Vrelativa = VLB - VLA
    # Calculando a velocidade relativa.
    tabela.add_row([angulo,e,Va,Vb,round(VLA,6),round(VLB,6),round(Vrelativa,6),Energia,round(inicio - fim,6)])
    # Adicionando os valores encontrados na tabela.
    Angulo.append(angulo)
    va_Linha.append(VLA)
    vb_Linha.append(VLB)
    v_relativa.append(Vrelativa)
    perda_de_energia.append(Energia)
    # Adicionando valores para montar o gráfico.
    angulo += 5
    # Para avançar para a próxima iteração. 
velocidades(massaa,massab,Va,Vb,angulo,e)
# Chamada do método para calcular as velocidades.
print(tabela)
# Apresentar tabela com os valores encontrados.
```

```
import numpy as np
import matplotlib.pyplot as plt
# importando bibliotecas necessárias.
plt.style.use('classic')
plt.figure(figsize=(8, 4))
plt.plot( Angulo, v_relativa,'o-', color='blue', label = 'Vrelativa')
plt.title('Soluções obtidas')
plt.xlabel('Ângulo (graus)')
plt.ylabel('Velocidades relativas (m/s)')
plt.ylim()
plt.xlim()
plt.legend(loc='lower right')
plt.grid('True')
plt.show()
# Configurações usadas para criar o gráfico.
```
